**Q24.** Make a class `Date` with class method `from_string(cls, s)` to parse `"DD-MM-YYYY"` and return an object.

```python
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    @classmethod
    def from_string(cls, s):
        # Split "DD-MM-YYYY" into parts
        d, m, y = s.split("-")

        # Convert strings to integers and create object
        return cls(int(d), int(m), int(y))


# Example usage
d1 = Date.from_string("25-12-2025")
print(d1.day, d1.month, d1.year)
```

**Q27.** Write a simple decorator that prints `"Function started"` before running the function. Test it on a small function.

```python
# TODO: Write your solution for Q27 below

def deco(func):
    def wrapper():
        print("function started")
        func()
    return wrapper

@deco
def hello():
    print("Hello")

hello()
```

**Q29.** Make a decorator that prints the **name** of the function being called (use `func.__name__`).

```python
# TODO: Write your solution for Q29 below

def show_name(func):
    def wrapper():
        print("Calling function:", func.__name__)
        func()
    return wrapper


@show_name
def hello():
    print("Hello!")


# Test
hello()
```

**Q30.** Combine two decorators: one prints `"Start"`, another prints `"End"` around a function. Check the order of messages.

```python
# TODO: Write your solution for Q30 below
def start_decorator(func):
    def wrapper():
        print("Start")
        return func()
    return wrapper


def end_decorator(func):
    def wrapper():
        result = func()
        print("End")
        return result
    return wrapper


@start_decorator
@end_decorator
def greet():
    print("Hello!")


# Test
greet()

```
