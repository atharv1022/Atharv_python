### 36. Determine the type of triangle using sides: acute, right, obtuse.

```python
a = int(input("Enter first side: "))
b = int(input("Enter second side: "))
c = int(input("Enter third side: "))

sides = sorted([a, b, c])
a, b, c = sides[0], sides[1], sides[2]


if c**2 == a**2 + b**2:
    print("Right triangle")
elif c**2 < a**2 + b**2:
    print("Acute triangle")
else:
    print("Obtuse triangle")
```

### 42. Check if a number is perfect (sum of factors equals number).

```python
sum_of_factors = 0
for i in range(1, num):
    if num % i == 0:
        sum_of_factors += i

if sum_of_factors == num:
    print("Perfect number")
else:
    print("Not a perfect number")
```

### 48. Check if a password contains both letters and digits.

```python
password = input("Enter password: ")

has_letter = False
has_digit = False

for ch in password:
    if ch.isalpha():
        has_letter = True
    if ch.isdigit():
        has_digit = True

if has_letter and has_digit:
    print("Valid: contains both letters and digits")
else:
    print("Invalid: must contain letters and digits")
```
