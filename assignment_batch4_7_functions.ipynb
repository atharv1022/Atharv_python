### What are positional arguments? Give an example.

```python
'''Positional arguments are values passed to a function or method based on their order or position in the function call. 
    The first argument provided in the call is assigned to the first parameter in the function's definition, 
    the second argument to the second parameter, and so on '''

def info(name,age):
    print("my name is ",name,"my age is",age)

info("atharv",23)
info("ptatham",25)
```

### What are keyword arguments? Give an example.

```python
'''Keyword arguments are values passed to a function that are identified by their parameter name, making the function call
more explicit and readable. This allows you to pass arguments in any order because the function uses the parameter name to assign
the value, not its position   '''

def info(name,city ,age):
    print("name:",name)
    print("city:",city)
    print("age:",age)

info(city = "kop",name = "atharv",age = 35)
```

### What does *args mean in Python? Write a function that accepts any number of values using *args and prints them.

```python
''' *args is a special syntax used in function definitions to allow a function to accept a variable number of positional arguments. 
The asterisk (*) before the parameter name (conventionally args) indicates that all positional arguments passed to the function will
be collected into a tuple. This tuple can then be iterated over or accessed by index within the function body.
-- input is tuple type '''

def val(*args):
    for val in args:
        print(val)

val(1,2,3,4)
val()
```

### What does **kwargs mean? Write a function that prints all keyword arguments using **kwargs.

```python
''' **kwargs is  used in function definitions to allow a function to accept an arbitrary number of keyword arguments.
The double asterisk (**) before a parameter name indicates that this parameter will collect all keyword arguments passed to the function into a 
dictionary. The keys of this dictionary will be the names of the keyword arguments, and the values will be their corresponding values.  
-- input is dictonary type '''

def val(**kwargs):
    for key,value in kwargs.items():
        print(f"{key} : {value}")


val(name="Alice", age=30, city="New York")
val(product="Laptop", price=1200.50)
```

### Write an adapter function that takes a list of operations and executes add, sub, mul, div based on input. 

### Example: [ {"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5} ]

```python


    
    
def calculator(**kwargs):
    if kwargs["op"]=="add":
        addition(kwargs["a"],kwargs["b"])
    elif kwargs["op"]=="sub":
        substraction(kwargs["a"],kwargs["b"])
    elif kwargs["op"]=="mul":
        multipy(kwargs["a"],kwargs["b"])
    elif kwargs["op"]=="div":
        divide(kwargs["a"],kwargs["b"])
    else:
        print("Provided operation is not exist")
        
        
def addition(x,y):
    print("sum=",x+y)

def substraction(x,y):
    print("sub=",x-y)

def multipy(x,y):
    print("mul=",x*y)

def divide(x,y):
    print("div=",x/y)
    

dic=[ {"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5},{"op": "sub", "a": 4, "b": 5} ,{"op": "div", "a": 4, "b": 5},{"op":"rem"}]
for i in dic:
    calculator(**i)
```

### Write out_of_stock(items) that returns all products with stock < 20.

```python
items = [
    {"name": "Laptop", "price": 60000, "stock": 15},
    {"name": "Mouse", "price": 500, "stock": 150},
    {"name": "Keyboard", "price": 1500, "stock": 8},
    {"name": "Monitor", "price": 12000, "stock": 25},
    {"name": "USB Cable", "price": 150, "stock": 50},
    {"name": "Headphones", "price": 2000, "stock": 18}
]

def out_of_stock(items):
    out_list = []
    for i in items:
        if i["stock"] <20:
            out_list.append(i["name"])

    print("list contain out of stock products:", out_list)

out_of_stock(items)
```
