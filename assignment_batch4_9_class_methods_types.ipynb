## Q1: Create a class called Student

Requirements:
- Class variable: `school_name = "ABC International School"`
- Constructor: takes `name` and `marks`
- Instance method: `display_details()`
- Class method: `change_school()`
- Static method: `is_pass(marks)` returns "Pass" if marks â‰¥ 35 else "Fail"


```python
class Student:
    school_name= "ABC International School"

    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
    
    def display_details(self):       #instance method
        print(self.name)
        print(self.marks)
        print(Student.school_name)
        print(Student.is_pass(marks=self.marks))

        
    @classmethod
    def change_school(cls,new_name):
        cls.school_name = new_name
        


    @staticmethod
    def is_pass(marks):
        if marks >= 35:
            return "pass"
        else:
            return "fail"

s1 = Student("Atharv",55)
s1.display_details()

Student.change_school("XYZ International School")
s1.display_details()

```

## Q2: Create a class Calculator

Requirements:
- Class variable: `count_operations = 0`
- Instance method: `add(a, b)`
- Class method: `increment_count()`
- Static method: `info()`


```python
class Calculator:
    count_operations = 0

    

    def add(self,a,b):
        result = a+b
        Calculator.increment_count()
        return result

    @classmethod
    def increment_count(cls):

        cls.count_operations +=1

    def info():
        print("this is a simple calculator")


c1 = Calculator()
print(c1.add(10,15))
print(c1.add(20,40))
print(Calculator.count_operations)
Calculator.info()
```

```python
class Calculator:
    # Class variable
    count_operations = 0

    # Instance Method
    def add(self, a, b):
        result = a + b
        Calculator.increment_count()  # increment count whenever add() is used
        return result

    # Class Method
    @classmethod
    def increment_count(cls):
        cls.count_operations += 1

    # Static Method
    @staticmethod
    def info():
        return "This is a simple calculator class."


c = Calculator()

print(c.add(10, 20))     # 30
print(c.add(5, 15))      # 20

print("total operations:",Calculator.count_operations)  # 2

print(Calculator.info())  # Info about class
```
