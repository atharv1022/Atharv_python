## **2. Find the sum of all numbers inside a nested list.**
Example: `[[10,20], [30], [5, 7, 8]]`

````python
# Write your answer here
lst=[[10,20], [30], [5, 7, 8]]
sum = 0
for i in lst:
    for j in i:
        sum+=j

print(sum)
````

## **8. Flatten a nested list into a single list.**
`[[1,2], [3,4], [5]] â†’ [1,2,3,4,5]`

````python
lst=[[1,2], [3,4], [5]]
lst1 = []
for sublst in lst:
    for value in sublst:
        lst1.append(value)

print(lst1)
````

## **9. Replace all negative numbers with 0 inside a nested list.**

````python
# Write your answer here
lst = [[80, -90, 85], [-75, 70, 65], [88, 92, -79]]
for i in range (len(lst)):
    for j in range(len(lst[i])):
        if lst[i][j]<0:
            lst[i][j] = 0

print(lst)
````

## **20. Convert all inner dictionary keys to uppercase.**

````python
employees = {
    'E1': {'name': 'John', 'status': 'active'},
    'E2': {'name': 'Sara', 'status': 'inactive'},
    'E3': {'name': 'Mike', 'status': 'active'},
    'E4': {'name': 'Rita', 'status': 'inactive'}
}

for emp_id,emp_info in employees.items():
    new_info = {}
    for key,val in emp_info.items():
        new_info[key.upper()]  = val
    employees[emp_id] = new_info

print(employees)
````
